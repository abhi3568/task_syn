```python
import pandas as pd
import numpy as np
import requests as rq
```

```python
# r=requests.get('http://ec2-52-90-42-239.compute-1.amazonaws.com:8080/SpringJDBCApp-0.0.1-SNAPSHOT/stocks/getstocksbyTicker?Ticker=AAPL&startDate=2015-11-01&endDate=2020-12-04')
data = rq.get('http://ec2-52-90-42-239.compute-1.amazonaws.com:8080/SpringJDBCApp-0.0.1-SNAPSHOT/stocks/getstocksbyTicker?Ticker=AAPL&startDate=2015-11-01&endDate=2020-12-04')
df = pd.DataFrame(data.json())
df=df['unsortedList']
```

```python
df1 = pd.DataFrame(df.values.tolist(), index=df.index)
df1
```

```python
df1['year']=pd.DatetimeIndex(df1['date']).year
df1['month']=pd.DatetimeIndex(df1['date']).month
df1['day']=pd.DatetimeIndex(df1['date']).day
```

```python
df1
```

```python
data1=pd.Series(df1['month'])
data2=pd.Series(df1['month']).shift(1)
```

```python
data1
```

```python
data2
```

```python
result=data1 != data2
```

```python

df1=df1[result.values]
```

```python
dt=rq.get('http://ec2-52-90-42-239.compute-1.amazonaws.com:8080/SpringJDBCApp-0.0.1-SNAPSHOT/stocks/dividend?Ticker=AAPL')
```

```python
dt
# new_data = pd.DataFrame(dt.json())
```

```python
# dt.loc[(dt['fiscal_period'] =='Q1')]
new_raw=dt.loc[(dt['fiscal_period'].isin(['Q1','Q2','Q3']))]
new_raw
# isin(['one','two'])]
```

```python
filter_date=new_raw.loc[(new_raw['end_date'] >='2015-12-31' )]
```

```python
filter_date
```

```python
by_desc=filter_date.sort_values(by='end_date', ascending=False)
by_desc
```

```python
by_desc['year']=pd.DatetimeIndex(by_desc['end_date']).year
by_desc['month']=pd.DatetimeIndex(by_desc['end_date']).month
by_desc['day']=pd.DatetimeIndex(by_desc['end_date']).day
```

```python
by_desc
```

```python
by_desc['month']
data_prev=pd.Series(by_desc['month'])
data_next=pd.Series(by_desc['month']).shift(1)
```

```python
result_raw=data_prev != data_next
```

```python
by_desc=by_desc[result_raw.values]
```

```python
by_desc
```

```python
df1
```

```python
new_df=pd.merge(df1, by_desc, how ='outer', on ='month')
```
